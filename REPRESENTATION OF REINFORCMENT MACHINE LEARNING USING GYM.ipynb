```python
import gym
import matplotlib.pyplot as plt


env =gym.make("Taxi-v3", render_mode="rgb_array").env
env.reset()
ggg = env.render()
env.action_space
env.observation_space
env.P
plt.imshow(ggg)
plt.show()
```

```python
epochs = 0
penalities = 0
done = 0
frame = []
while not done:
    action = env.action_space.sample()
    state, reward, done, info, oot= env.step(action)
    if reward == -10:
        penalities +=1
    frame.append({"frame": env.render(),
                 "state": state,
                  "action": action,
                  "reward": reward,
                  "done": done,
                  "info": info
                  })
    epochs += 1
print(epochs)
print(penalities)
```

```python
from IPython.display import clear_output
from time import sleep
def displayFrame(frame):
    for i, frame in enumerate(frame):
        clear_output(wait = True)
        hhh = frame["frame"]
        plt.imshow(hhh)
        plt.show()
        
        print(frame["state"])
        print(frame["action"])
        print(frame["reward"])
        print(i)
        sleep(0.01)
```
